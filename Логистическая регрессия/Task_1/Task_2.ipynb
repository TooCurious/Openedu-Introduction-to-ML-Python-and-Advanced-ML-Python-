# Логистическая регрессия (Заданий 1)
В прилагаемом файле представлены данные, собранные путем голосования за самые лучшие (или, по крайней мере, самые популярные) конфеты Хэллоуина. Обучите модель логистической регрессии. В качестве предикторов выступают поля: **chocolate**, **fruity**, **caramel**, **peanutyalmondy**, **nougat**, **crispedricewafer**, **hard**, **bar**, **pluribus**, **sugarpercent**, **pricepercent**, отклик — **Y**.

В качестве тренировочного набора данных используйте данные из файла, за иключением следующих конфет: **Candy Corn**, **Nerds**, **Reeses stuffed with pieces**. Обучите модель.

Обучите модель и выполните предсказание для всех конфет из прилагаемого файла тестовых данных.

1. Введите вероятность отнесения конфеты Tootsie Roll Juniours к классу 1.
2. Введите вероятность отнесения конфеты Swedish Fish к классу 1.
3. Выполните оценку модели с помощью матрицы ошибок и рассчитайте следующие параметры при пороге отсечения (Treshhold).
4. Введите значение Recall, или TPR для тестового набора данных.
5. Введите значение Precision для тестового набора данных.
6. Введите значение AUC для тестового набора данных.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from pandas import DataFrame

# тренировочные данные
data_train = pd.read_csv('/content/drive/MyDrive/candy-data.csv', index_col='competitorname')
data_train.drop(['winpercent'], axis=1, inplace=True)
data_train.drop(['Candy Corn', 'Nerds', 'Reeses stuffed with pieces'], axis=0, inplace=True)
y = data_train.loc[:, 'Y'].to_numpy()
data_train.drop(['Y'], axis=1, inplace=True)
X = data_train.to_numpy()

# тестовые данные
data_test = pd.read_csv('/content/drive/MyDrive/candy-test.csv', index_col='competitorname')

Tootsie_Roll_Juniors = data_test.loc['Tootsie Roll Juniors'].drop(labels='Y').to_numpy().reshape(1, -1)
Swedish_Fish = data_test.loc['Swedish Fish'].drop(labels='Y').to_numpy().reshape(1, -1)


reg = LogisticRegression(random_state=2019, solver='lbfgs').fit(X, y)

print(f'Tootsie_Roll_Juniors = {np.around(*reg.predict_proba(Tootsie_Roll_Juniors)[:,1], 3)}')
print(f'Swedish_Fish = {np.around(*reg.predict_proba(Swedish_Fish)[:,1], 3)}')

con_mat = confusion_matrix(reg.predict(data_test.loc[:, :'pricepercent'].to_numpy()),
                       data_test.loc[:, 'Y'].to_numpy().reshape(-1, 1))

AUC_ROC = np.around(roc_auc_score(data_test.loc[:, 'Y'].to_numpy().reshape(-1, 1),
                              reg.decision_function(data_test.loc[:, :'pricepercent'].to_numpy())), 3)

print(f'Матрица ошибок [[TP FP] [FN TN]] {con_mat}')
TPR = np.around(con_mat[1, 1] / (con_mat[1, 1] + con_mat[0, 1]), 3)
Precision = np.around(con_mat[1, 1] / (con_mat[1, 1] + con_mat[1, 0]), 3)
print(f'TPR = {TPR}')
print(f'Precision = {Precision}')
print(f'AUC_ROC = {AUC_ROC}')
```
