2. По результатам выполнения кластеризации определить среднее расстояний между объектами и центроидом, отнесенных к кластеру 0.

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from sklearn.cluster import KMeans

data_frame = pd.read_csv('/content/drive/MyDrive/data.csv', index_col='Object')
data_frame.drop(['Cluster'], axis=1, inplace=True)

kmeans = KMeans(n_clusters=3, init=np.array([[12.2, 11.4], [15.75, 12.25], [13.67, 9.67]]), max_iter=100, n_init=1).\
    fit(data_frame)

center_array = kmeans.cluster_centers_[0]

data_frame['Cluster'] = kmeans.labels_
df0 = data_frame[data_frame['Cluster'] == 0]

dist = 0
for i in [1, 6, 15]:
    dist += ((center_array[0] - df0.loc[i, 'X']) ** 2 + (center_array[1] - df0.loc[i, 'Y']) ** 2) ** 0.5
print(f'Среднее расстояние между объектами и центройдом для класса 0 = {np.around(dist / df0.shape[0], 3)}')
print(data_frame)
```
